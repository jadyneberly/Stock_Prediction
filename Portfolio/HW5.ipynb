# Dimensionality Reduction

## Instructions

**Option 1**

* Using SP500 stocks, load historical values of the stocks SP500 firms (5 pts)
* Create cumulative future returns of the target variable and save it in the **variable Y**. You cannot choose MSFT or AAPL as the target variable (5 pts)
* Create cumulative returns of the other stocks and save them in the **variable X** (5 pts)
* Split X and Y into train-test samples (2.5 pts)
* Create a pipeline (10 pts)
    * Add a step that handles missing values
    * Add a step that standardizes the scale of the features
    * Add a step that keeps the 3 most informative features
    * Add a step that utilizes Linear Regression
* Create a pipeline (10 pts)
    * Add a step that handles missing values
    * Add a step that standardizes the scale of the features
    * Add a step that reduces dimensions using PCA and a threshold variance of 90%
    * Add a step that utilizes Lasso Regression
* Create a pipeline (10 pts)
    * Add a step that handles missing values
    * Add a step that standardizes the scale of the features
    * Add a step that reduces dimensions using Kernel PCA and a threshold variance of 90%
    * Add a step that utilizes Lasso Regression
* Compare the three pipelines using K-fold results (7.5 pts)
* Perform model tuning on the 3rd pipeline (10 pts)
    * At least two parameters must vary
* Save the model (5 pts)
* Rank the features by importance (5 pts)
* Save the SHAP Explainer and create a SHAP plot (5 pts)
* Deploy the model on AWS (10 pts)
* Deploy the Streamlit app (10 pts)
